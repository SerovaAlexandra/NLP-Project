```python
import pandas as pd
```

```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
from nltk.tokenize import word_tokenize
```

```python
data = []
```

```python
def tag_handler(pos_tag):
    if pos_tag != None:
        if pos_tag.startswith('ADJ'):
            pos_tag = 'ADJ'
            return pos_tag
        elif pos_tag == 'INFN':
            pos_tag = 'VERB'
            return pos_tag
        elif pos_tag.startswith ('PRT'):
            pos_tag = 'PRT'
            return pos_tag
        else:
            return pos_tag
    else:
        return "None"
```

```python
from natasha import (
    Segmenter,
    MorphVocab,
    
    NewsEmbedding,
    NewsMorphTagger,
    NewsSyntaxParser,
    NewsNERTagger,
    
    PER,
    NamesExtractor,

    Doc
)


segmenter = Segmenter()
morph_vocab = MorphVocab()

emb = NewsEmbedding()
morph_tagger = NewsMorphTagger(emb)
syntax_parser = NewsSyntaxParser(emb)
ner_tagger = NewsNERTagger(emb)

names_extractor = NamesExtractor(morph_vocab)
```

```python
with open ('emu_bilo_tiho.txt', encoding='utf-8') as f1:
    text = f1.read()
doc1 = Doc (text)
```

```python
doc1.segment(segmenter)
for elem1 in doc1.sents:
    sent_data_text1 = {}
    lemma_tag_data = []
    sent_text = elem1.text
    if '\n' in sent_text:
        sent_text = sent_text.replace ('\n', ' ')
    tokenizer = word_tokenize(sent_text)
    for w in tokenizer:
        if w != None:
            lemm_data = []
            lemm = morph.parse(w)[0].normal_form
            lemm_data.append(lemm)
            pos_data = tag_handler(morph.parse(w)[0].tag.POS)
            lemm_data.append(pos_data)
            lemma_tag_data.append(lemm_data)
    sent_data_text1['sent_text'] = sent_text
    sent_data_text1 ['lemmas_and_tags'] = lemma_tag_data
    sent_data_text1 ['link'] = 'https://ficbook.net/readfic/1580084'
    sent_data_text1 ['author'] = 'Чацкая'
    sent_data_text1 ['name'] = 'Ему было тихо'
    data.append(sent_data_text1)
```
